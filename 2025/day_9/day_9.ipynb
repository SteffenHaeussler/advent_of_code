## Day 9

### Test

```python
test = """7,1
11,1
11,7
9,7
9,5
2,5
2,3
7,3
"""

test = test.split("\n")[:-1]
test = [i.split(",") for i in test]
test= [[int(j) for j in i] for i in test]
```

```python
def calc_area(p1, p2):

    return (abs(p1[0]-p2[0])+1)*(abs(p1[1]-p2[1])+1)


def get_area_heap(grid):

    area_heap = []
    
    for i in range(len(grid)-1):
        for j in range(i+1, len(grid)):
            heapq.heappush(area_heap, [-calc_area(grid[i], grid[j]), (grid[i], grid[j])])
            
    return area_heap


def get_n_largest(area_heap, n=1):

    return heapq.nsmallest(1, area_heap)[0]
```

```python
area_heap = get_area_heap(test)
result, (_, _) = get_n_largest(area_heap)
```

```python
assert -result == 50
```

### Test

```python
test = """7,1
11,1
11,7
9,7
9,5
2,5
2,3
7,3
"""

test = test.split("\n")[:-1]
test = [i.split(",") for i in test]
test= [[int(j) for j in i] for i in test]
```

```python
def get_line(x1,y1,x2,y2):

    line = set()

    x_min, x_max = sorted([x1, x2])
    y_min, y_max = sorted([y1, y2])
    
    for x in range(x_min, x_max+1):
        for y in range(y_min, y_max+1):
            line.add((x,y))

    return line
    
    
def get_perimeter(coords):
    perimeter = set()

    for i in range(1, len(coords)):
        line = get_line(*coords[i-1], *coords[i])
    
        perimeter |= line
    
    line = get_line(*coords[-1], *coords[0])
    perimeter |= line

    return perimeter


def inside_perimeter(perimeter, x1, y1, x2, y2):
    x_min, x_max = sorted([x1, x2])
    y_min, y_max = sorted([y1, y2])

    for x_p, y_p in perimeter:
        if (x_min < x_p < x_max) and (y_min < y_p < y_max):
            return False
    return True


def get_max_area(perimeter, area_heap):

    while True:
    
        area, (p1, p2) = heapq.heappop(area_heap)
    
        if inside_perimeter(perimeter, *p1, *p2):
            return -area

    return False
```

```python
perimeter = get_perimeter(test)
area_heap = get_area_heap(test)
result = get_max_area(perimeter, area_heap)
```

```python
assert result == 24
```
