```python
from collections import defaultdict

import time
```

## Day 6

```python
def parse_data(data, from_left=True):
    arrays = defaultdict(list)

    lines = data.split("\n")
    if not lines[-1]:
        lines.pop()
        
    ops = lines.pop().split()

    if from_left:
    
        for line in lines:
            for i, val in enumerate(line.split()):
                arrays[i].append(int(val))

    else:
        temp = defaultdict(list)
        
        for line in lines:
            for i, val in enumerate(line):
                temp[i].append(val)

        cursor = 0
        
        for i in sorted(temp.keys()):
            val = "".join(temp[i])
            
            if val.strip():
                arrays[cursor].append(int(val))
            else:
                cursor += 1

    
    return ops, arrays
```

```python
def calculate(ops, arrays):

    results = []
    
    for i, op in enumerate(ops):
    
        if op == "+":
            results.append(sum(arrays[i]))
        else:
            temp = 1
            for val in arrays[i]:
                temp *= val
            results.append(temp)

    return results
```

### Test

```python
test = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
"""
```

```python
ops, arrays = parse_data(test)
results = calculate(ops, arrays)
```

```python
assert sum(results) == 4277556
```
