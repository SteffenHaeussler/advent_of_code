## Day 4

### Test

```python
test = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
"""
```

```python
test = test.split("\n")[:-1]
test = [list(i) for i in test]
```

```python
def get_paper_positions(grid):

    positions = []
    cols = len(grid[0])
    rows = len(grid)
    
    for row in range(rows):
        for col in range(cols):
    
            if grid[row][col] == ".":
                continue
    
            cnt = 0
    
            for i in range(row-1,row+2,1):
                for j in range(col-1,col+2,1):
            
                    if i < 0 or i >= rows or j < 0 or j >= cols or (i == row and j == col):
                        continue
    
                    if grid[i][j] == "@":
                        cnt += 1
    
            if cnt < 4:
                positions.append([row,col])

    return positions
```

```python
positions = get_paper_positions(test)
```

```python
assert len(positions) == 13
```

### Test

```python
test = """..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
"""
```

```python
test = test.split("\n")[:-1]
test = [list(i) for i in test]
```

```python
step = 0
positions = []
n_paper_rolls = 0

while True:

    positions = get_paper_positions(test)

    if not positions:
        break

    n_paper_rolls += len(positions)

    for row, col in positions:
        test[row][col] = "."

    step += 1
```
