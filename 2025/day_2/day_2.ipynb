## Day 2

### Test

```python
test = """11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"""
```

```python
test = test.split(",")
test = [i.split("-") for i in test]
```

```python
def is_invalid(s):

    n = len(s)
    
    if n % 2 == 1:
        return False

    start = 0
    mid = n // 2

    for i in range(start, mid):
        if s[i] != s[mid+i]:
            return False

    return True
```

```python
invalid = []

for start, end in test:

    for i in range(int(start), int(end)+1):

        if is_invalid(str(i)):
            invalid.append(int(i))
```

```python
result = sum(invalid)
```

```python
assert result == 1227775554
```

## Part 2

### Test

```python
test = """11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"""
```

```python
test = test.split(",")
test = [i.split("-") for i in test]
```

```python
def invalid_combination(s):

    for n_split in range(1, len(s)):

        splits = list(batched(s, n_split))

        if len(splits) == 1:
            if len(set(splits[0])) == 1:
                return True

        if all(i == splits[0] for i in splits[1:]): 
            return True

    return False
```

```python
invalid = []

for start, end in test:

    for i in range(int(start), int(end)+1):

        if invalid_combination(str(i)):
            
            invalid.append(i)
```

```python
result = sum(invalid)
```

```python
assert result == 4174379265
```
