## Part 2

### Test

```python
test = """L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
"""
```

```python
test = test.split("\n")[:-1]
test = [[i[0],int(i[1:])] for i in test]
```

```python
count = 0
start = 50

for rotation, distance in test:

    if rotation == "L":
        
        start -= distance

        a, b = divmod(start,100)

        if (start + distance) == 0:
            a += 1

        count += abs(a)

        if b == 0:
            count += 1

        start = b
        
    else:
        start += distance

        a, b = divmod(start,100)

        count += abs(a)
        
        start = b
```

```python
assert count == 6
```
