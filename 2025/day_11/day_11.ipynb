```python
from functools import cache

import time
```

## Day 11

### Test

```python
test = """aaa: you hhh
you: bbb ccc
bbb: ddd eee
ccc: ddd eee fff
ddd: ggg
eee: out
fff: out
ggg: out
hhh: ccc fff iii
iii: out
"""

test = test.split("\n")[:-1]
```

```python
def get_graph(grid):

    grid = [i.split(":") for i in grid]
    d = {}

    for i in grid:
        d[i[0]] = i[1].strip().split(" ")

    return d


@cache
def count_paths(start, target):
    
    if start == target:
        return 1

    return sum(count_paths(node, target) for node in d.get(start, []))
```

```python
start = "you"
target = "out"

d = get_graph(test)
results = count_paths(start, target)

count_paths.cache_clear()
```

```python
assert results == 5
```

```python
with open("../../../advent_of_code_input/2025/day_11/input.txt", "r") as f:
    data = f.read()

data = data.split("\n")[:-1]
```

```python
start_time = time.perf_counter()

start = "you"
target = "out"

d = get_graph(data)
results = count_paths(start, target)

end_time = time.perf_counter()

elapsed = end_time - start_time
print(elapsed, "seconds")

count_paths.cache_clear()
```
