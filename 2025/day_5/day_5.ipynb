## Part 2

### Test

```python
test = """3-5
10-14
16-20
12-18

1
5
8
11
17
32
"""
```

```python
ranges, _ = test.split("\n\n")

ranges = ranges.split("\n")

ranges = [i.split("-") for i in ranges]
ranges = [[int(j) for j in i] for i in ranges]
```

```python
def add_ingredients(available, start, end):

    idx_low = bisect.bisect_left(available, [start,end])
    
    if idx_low > 0 and available[idx_low-1][1] >= start:
        idx_low -= 1
        start = min(start, available[idx_low][0])
        end = max(end, available[idx_low][1])

    idx_upp = idx_low

    while idx_upp < len(available) and available[idx_upp][0] <= end:
        end = max(end, available[idx_upp][1])
    
        idx_upp += 1

    available[idx_low:idx_upp] = [[start,end]]

    return available
```

```python
count = 0
available = []

for start, end in ranges:
    available = add_ingredients(available, start,end)

for start, end in available:
    count += (end-start+1)
```

```python
assert count == 14
```
