## Day 12

### Test

```python
test = """0:
###
##.
##.

1:
###
##.
.##

2:
.##
###
##.

3:
##.
###
##.

4:
###
#..
###

5:
###
.#.
###

4x4: 0 0 0 0 2 0
12x5: 1 0 1 0 2 2
12x5: 1 0 1 0 3 2
"""

test = test.split("\n\n")
```

```python
def get_data(data):

    shapes = {}
    regions = []

    
    _shapes, _regions = data[:-1], data[-1]
    _shapes = [i.split(":\n") for i in _shapes]
    
    for i, shape in _shapes:
        shapes[int(i)] = shape.split("\n")

    _regions = _regions.split("\n")
    _regions = [i.split(": ") for i in _regions]


    for grid, indices in _regions[:-1]:
        row, col = grid.split("x")
        indices = indices.split(" ")
        indices = [(n, int(i)) for n, i in enumerate(indices) if i != "0"]
        regions.append((int(row), int(col), indices))

    return shapes, regions
```

```python
shapes, regions = get_data(test)
```

```python
n_shapes = {k: sum(i.count("#") for i in v) for k,v in shapes.items()}
```

```python
results = 0

for row, col, shape in regions:
    max_area = row*col
    res = 0
    for idx, amount in shape:
        res += amount * n_shapes[idx]

    if max_area >= res:
        results += 1
```

## Part 1

```python
with open("../../../advent_of_code_input/2025/day_12/input.txt", "r") as f:
    data = f.read()

data = data.split("\n\n")
```

```python
shapes, regions = get_data(data)
```

```python
n_shapes = {k: sum(i.count("#") for i in v) for k,v in shapes.items()}
```

```python
start_time = time.perf_counter()

results = 0

for row, col, shape in regions:
    max_area = row*col
    res = 0
    for idx, amount in shape:
        res += amount * n_shapes[idx]

    if max_area >= res:
        results += 1

end_time = time.perf_counter()

elapsed = end_time - start_time
print(elapsed, "seconds")
```
