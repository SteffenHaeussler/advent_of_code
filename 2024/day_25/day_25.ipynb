```python
test = """#####
.####
.####
.####
.#.#.
.#...
.....

#####
##.##
.#.##
...##
...#.
...#.
.....

.....
#....
#....
#...#
#.#.#
#.###
#####

.....
.....
#.#..
###..
###.#
###.#
#####

.....
.....
.....
#....
#.#..
#.#.#
#####"""
test = test.split("\n\n")
test = [i.split("\n") for i in test]
```

```python
def count(item):

    cnt = {i: 0 for i in range(5)}

    for line in item:
        for n, i in enumerate(line):
            if i == "#":
                cnt[n] += 1

    return list(cnt.values())
```

```python
def counter(items):

    locks = []
    keys = []

    for item in items:

        if "#" in item[0]:
            cnt = count(item[1:])
            locks.append(cnt)
        else:
            cnt = count(item[:-1])
            keys.append(cnt)

    return locks, keys
```

```python
def test_keys(locks, keys):

    result = 0
    
    for lock in locks:
        for key in keys:
            if all([i+j <=5 for i,j in zip(lock, key)]):
                result += 1

    return result
```

```python
locks, keys = counter(test)
result = test_keys(locks, keys)
```

```python
assert result == 3
```
